### Difference: 
#### Difference of the set B from set A(A - B) is a set of elements that are only in A but not in B. Similarly, B - A is a set of elements in B but not in A.

Syntax: (A-B)


```python
A = {10, 20, 30, 40, 80}
B = {100, 30, 80, 40, 60}
print(A.difference(B))
print(B.difference(A))
```

### Difference_update:
    
The set difference_update() method modifies the existing set. If (A – B) is performed, then A gets modified into (A – B), and if (B – A) is performed, then B gets modified into (B – A).

A.difference_update(B) for (A - B)
B.difference_update(A) for (B - A)



```python
A = {10, 20, 30, 40, 80}
B = {100, 30, 80, 40, 60}
#Modifies A and returns None
print(A.difference_update(B))
# Prints the modified set
print (A)
```

```python
#Modifies B and returns None
print(B.difference_update(A))
# Prints the modified set
print(B)
```

### Symmetric Difference:
- This in-built function of Python Set helps us to get the symmetric difference between two sets, which is equal to the elements present in either of the two sets, but not common to both the sets. Let’s look at the Venn diagram of the symmetric_difference between two sets.

#### Syntax: (A^B) or (B^A)

- A.symmetric_difference(B)
- B.symmetric_difference(A)

```python
set_A = {"ram", "rahim", "ajay", "rishav", "aakash"}
set_B = {"aakash", "ajay", "shyam", "ram", "ravi"}
set_A.symmetric_difference(set_B)
#  The common elements {3, 4} are discarded from the output.
```

```python
set_B.symmetric_difference(set_A)
```

### Symmetric_Difference_Update:
The symmetric_difference_update of two sets A and B is the set of elements that are in either A or B, but not in their intersection.

#### Syntax: A.symmetric_difference_update(B) 

- Symmetric_difference_update() method updates the set calling symmetric_difference_update() with the symmetric difference of sets.




Returns: 
- This method returns None (which indicates absence of a return value). It only updates the set calling symmetric_difference_update() with the symmetric difference of sets.

```python
set_A = {'p', 'a', 'w', 'a', 'n'}
set_B = {'r', 'a', 'o', 'n', 'e'}
print(set_A.symmetric_difference_update(set_B))
```

```python
print(set_B.symmetric_difference_update(set_A))
```

### isdisjoint()
- The isdisjoint() method returns True if two sets are disjoint sets. If not, it returns False.

- Two sets are said to be disjoint sets if they have no common elements. For example:

Syntax: set_a.isdisjoint(b)

```python
A = {1, 2, 3, 4}
B = {5, 6, 7}
C = {4, 5, 6}

print('Are A and B disjoint?', A.isdisjoint(B))
print('Are A and C disjoint?', A.isdisjoint(C))
```

### issubset()
The issubset() method returns True if all elements of a set are present in another set (passed as an argument). If not, it returns False.

Syntax: A.issubset(B)

```python
A = {1, 2, 3, 4,5,6,7}
B = {5, 6, 7}
C = {4, 5, 6}
print(A.issubset(B))
print(B.issubset(A))
print(C.issubset(A))
```

### issuperset()
The issuperset() method returns True if a set has every elements of another set (passed as an argument). If not, it returns False.

Set X is said to be the superset of set Y if all elements of Y are in X.

Syntax: A.issuperset(B)

```python
A = {1, 2, 3, 4,5,6}
B = {1, 2, 3}
C = {4, 5, 6}
print(A.issuperset(B))
print(B.issuperset(A))
print(A.issuperset(C))
```
