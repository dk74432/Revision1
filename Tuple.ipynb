#### Remove empty tuples from a list

```python
tuples = [(), ('ram','15','8'), (), ('laxman', 'sita'), 
                  ('krishna', 'akbar', '45'), ('',''),()]
def Remove(tup):
    tup = [i for i in tuples if i!=()]
    return tup
print(Remove(tuples))
```

#### Sort a tuple by its float element

```python
def Sort(tup):
    # reverse = True (Sorts in Descending order)
    # key is set to sort using float elements
    # lambda has been used
    return(sorted(tup, key = lambda x: float(x[1]), ))
  
# Driver Code
tup = [('lucky', '18.265'), ('nikhil', '14.107'), ('akash', '24.541'), 
    ('anand', '4.256'), ('gaurav', '10.365')]
print(Sort(tup))
```

```python
tuples= [('lucky', '18.265'), ('nikhil', '14.107'), 
                  ('akash', '24.541'), ('anand', '4.256'), ('gaurav', '10.365')]
Sort=sorted(tuples,key=lambda x:float(x[1]))
for i in Sort:
    print(i)
```

##### Python | Count the elements in a list until an element is a Tuple

```python
lst=[4, 5, 6, 10, (1, 2, 3), 11, 2, 4]
def count(lst):
    counter=0
    for i in lst:
        if isinstance(i,tuple):
            break
        counter=counter+1
    return counter
count(lst)
```
